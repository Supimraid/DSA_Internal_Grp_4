```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import requests
import json
```

```python
rows_before = len(df)

# Drop all rows that contain *any* null value in any column
df = df.dropna().reset_index(drop=True)
rows_after = len(df)

print(f"Rows before cleanup: {rows_before}")
print(f"Rows after cleanup:  {rows_after}")
print("\nNull counts after dropping rows with missing values (all columns):")
print(df.isnull().sum())
```
